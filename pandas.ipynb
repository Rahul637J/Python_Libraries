3. Write a Python program to add, subtract, multiple and divide two Pandas Series.

```python

def perform_operations(series1, series2):
    
    """
    Description:
        This function performs addition, subtraction, multiplication, and division 
        on two given Pandas Series.
    Parameters:
        series1 (pd.Series): The first Pandas Series.
        series2 (pd.Series): The second Pandas Series.
    Returns:
        dict: A dictionary containing the results of the four operations.
    """
    
    operations = {
        "Addition": series1 + series2,
        "Subtraction": series1 - series2,
        "Multiplication": series1 * series2,
        "Division": series1 / series2
    }
    return operations

def main():
    
    series1 = pan.Series([10, 20, 30, 40, 50])
    series2 = pan.Series([2, 4, 6, 8, 10])

    results = perform_operations(series1, series2)

    for operation, result in results.items():
        print(f"{operation}:\n{result}\n")

if __name__ == "__main__":
    main()
```

4. Write a Python program to get the powers of an array values element-wise.

```python

def compute_elementwise_powers(base_series, exponent_series):
    """
    Description:
        This function computes the element-wise powers of the base Series values raised to the values of the exponent Series.
    Parameters:
        base_series (pd.Series): The Series with base values.
        exponent_series (pd.Series): The Series with exponent values.
    Returns:
        pd.Series: A Series containing the result of raising each element of the base Series to the corresponding power in the exponent Series.
    """

    if len(base_series) != len(exponent_series):
        raise ValueError("Both Series must have the same length.")
    
    result = base_series ** exponent_series
    return result

def main():
    base_series = pan.Series([0, 1, 2, 3, 4, 5, 6])
    exponent_series = pan.Series([1, 2, 3, 4, 5, 6, 7])
    
    result = compute_elementwise_powers(base_series, exponent_series)
    
    print("Resulting Series with element-wise powers:\n", result)

if __name__ == "__main__":
    main()
```

```python

def append_row(df, row_label, row_data):
    
    """
    Description:
        Appends a new row to the DataFrame with the given label and data.
    Parameters:
        df (pd.DataFrame): The DataFrame to which the new row will be added.
        row_label (str): The label of the new row.
        row_data (dict): A dictionary containing the data for the new row.
    Returns:
        pd.DataFrame: The DataFrame with the new row added.
    """

    df.loc[row_label] = row_data
    return df

def delete_row(df, row_label):
    """
    Description:
        Deletes the specified row from the DataFrame.
    Parameters:
        df (pd.DataFrame): The DataFrame from which the row will be deleted.
        row_label (str): The label of the row to be deleted.
    Returns:
        pd.DataFrame: The DataFrame with the specified row deleted.
    """

    df = df.drop(row_label)
    return df

def create_dataframe(exam_data, labels):
    """
    Description:
        Creates a DataFrame from the given data and index labels.
    Parameters:
        exam_data (dict): The data to include in the DataFrame.
        labels (list): The index labels for the DataFrame.
    Returns:
        pd.DataFrame: The created DataFrame.
    """

    return pan.DataFrame(exam_data, index=labels)

def main():
    exam_data = {
        'name': ['Anastasia', 'Dima', 'Katherine', 'James', 'Emily', 'Michael', 'Matthew',
                 'Laura', 'Kevin', 'Jonas'],
        'score': [12.5, 9, 16.5, np.nan, 9, 20, 14.5, np.nan, 8, 19],
        'attempts': [1, 3, 2, 3, 2, 3, 1, 1, 2, 1],
        'qualify': ['yes', 'no', 'yes', 'no', 'no', 'yes', 'yes', 'no', 'no', 'yes']
    }
    labels = ['a', 'b', 'c', 'd', 'e', 'f', 'g', 'h', 'i', 'j']
    
    df = create_dataframe(exam_data, labels)
    
    print("Original DataFrame:")
    print(df)
    
    new_row_data = {
        'name': 'Suresh',
        'score': 15.5,
        'attempts': 1,
        'qualify': 'yes'
    }
    
    df = append_row(df, row_label='k', row_data=new_row_data)
    
    print("\nDataFrame after adding row 'k':")
    print(df)
    
    df = delete_row(df, row_label='k')
    
    print("\nDataFrame after deleting row 'k':")
    print(df)

if __name__ == "__main__":
    main()
```
