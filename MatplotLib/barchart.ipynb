5. Write a Python programming to display a bar chart of the popularity of programming
Languages. Attach a text label above each bar displaying its popularity (float value).

```python
def create_bar_chart(programming_language,popularity):
    
    """
    Description:
        Creates and displays a vertical bar chart comparing the popularity of different programming languages.
        Each bar is labeled with its corresponding popularity value, and the bars are displayed with labels for clarity.

    Parameters:
        programming_language (list of str): A list of programming languages to be plotted on the x-axis.
        popularity (list of float): A list of popularity percentages corresponding to each programming language to be plotted on the y-axis.

    Return Type:
        None
    """

    for i in range(len(programming_language)):
        mp.bar(programming_language[i],popularity[i],label=popularity[i])
        mp.text( i,popularity[i] , str(popularity[i]),fontsize=12, ha='center', va='bottom')

    mp.tight_layout()

    mp.show()


def main():

    programming_languages= ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
    popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
    create_bar_chart(programming_languages,popularity)

if __name__=="__main__":
    main()
```

Languages. Make blue border to each bar.

```python
def create_bar_chart(programming_language,popularity):
    
    """
    Description:
        Creates and displays a vertical bar chart comparing the popularity of different programming languages.
        Each bar is displayed with its popularity value annotated on top. The chart includes gridlines for better readability.

    Parameters:
        programming_language (list of str): A list of programming languages to be plotted on the x-axis.
        popularity (list of float): A list of popularity percentages corresponding to each programming language to be plotted on the y-axis.

    Return Type:
        None
    """

    for i in range(len(programming_language)):
        mp.bar(programming_language[i],popularity[i],edgecolor="blue")
        mp.text( i,popularity[i] , str(popularity[i]), ha='center', va='bottom')

    mp.tight_layout()

    
    mp.grid(True,which="major",color="blue")
    mp.minorticks_on()  
    mp.grid(True,which="minor",)

    mp.show()


def main():

    programming_languages= ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
    popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
    create_bar_chart(programming_languages,popularity)

if __name__=="__main__":
    main()
```

7. Write a Python programming to display a bar chart of the popularity of programming
Languages. Specify the position of each bar plot.

```python
def create_bar_chart(programming_language,popularity):
    
    """
    Description:
        Creates and displays a vertical bar chart comparing the popularity of different programming languages.
        Each bar is positioned based on an array of positions, with its popularity value annotated on top. 
        The chart uses blue for bar edges and includes positions for better alignment.

    Parameters:
        programming_language (list of str): A list of programming languages to be plotted on the x-axis.
        popularity (list of float): A list of popularity percentages corresponding to each programming language to be plotted on the y-axis.

    Return Type:
        None
    """

    positions=np.arange(1,len(programming_language)+1) # This will return the array of positions from 1
    mp.bar(positions,popularity,edgecolor="blue")

    for i in range(len(programming_language)):
        mp.text(positions[i], popularity[i], str(popularity[i]), ha='center', va='bottom')

    mp.show()

def main():

    programming_languages= ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
    popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
    create_bar_chart(programming_languages,popularity)

if __name__=="__main__":
    main()
```

8. Write a Python programming to display a bar chart of the popularity of programming
Languages. Select the width of each bar and their positions.

```python
def create_bar_chart(programming_language,popularity):

    """
    Description:
        Creates and displays a vertical bar chart comparing the popularity of different programming languages.
        Each bar's height represents the popularity value, and the bars are annotated with their respective values.
        The x-axis labels correspond to the programming languages, and the bars are displayed with blue edges.

    Parameters:
        programming_language (list of str): A list of programming languages to be plotted on the x-axis.
        popularity (list of float): A list of popularity percentages corresponding to each programming language to be plotted on the y-axis.

    Return Type:
        None
    """
    
    positions=np.arange(1,len(programming_language)+1)

    for i in range(len(programming_language)):
        mp.bar(positions,popularity,edgecolor="blue",width=1)
        mp.text(positions[i], popularity[i] , str(popularity[i]), ha='center', va='bottom')

    mp.xticks(positions,programming_language)    

    mp.show()

def main():

    programming_languages= ["Java", "Python", "PHP", "JavaScript", "C#", "C++"]
    popularity=[22.2, 17.6, 8.8, 8, 7.7, 6.7]
    create_bar_chart(programming_languages,popularity)

if __name__=="__main__":
    main()
```

10. Write a Python program to create bar plot of scores by group and gender. Use multiple X
values on the same chart for men and women.

```python
def create_bar_chart(means_men,means_women):
    
    """
    Description:
        Creates and displays a grouped bar chart comparing the scores of two groups (Men and Women) across multiple categories.
        The chart visually represents the scores of both groups side by side for each category, allowing for easy comparison.

    Parameters:
        means_men (tuple of float): A tuple containing the average scores for the 'Men' group across different categories.
        means_women (tuple of float): A tuple containing the average scores for the 'Women' group across different categories.

    Return Type:
        None
    """

    n = len(means_men)  
    
    ran = np.arange(n)  
    width = 0.35 
    
    fig, ax = mp.subplots(figsize=(10, 7))
    
    ax.bar(ran - width/2, means_men, width, label='Men', color='blue')
    
    ax.bar(ran + width/2, means_women, width, label='Women', color='red')
    
    ax.set_xlabel('Group')
    ax.set_ylabel('Scores')
    ax.set_title('Scores by group and gender')
    ax.set_xticks(ran)
    ax.set_xticklabels(['Group 1', 'Group 2', 'Group 3', 'Group 4', 'Group 5'])
    
    ax.legend()

    mp.show()

def main():

    means_men = (22, 30, 35, 35, 26)
    means_women = (25, 32, 30, 35, 29)
    create_bar_chart(means_men,means_women)

if __name__=="__main__":
    main()
```

11. Write a Python program to create bar plot from a DataFrame.

```python

def create_dataframe(data):
    """
    Description:
        Creates and returns a Pandas DataFrame from the provided data.

    Parameters:
        data (dict of lists): A dictionary where the keys are column names and the values are lists of column data.

    Return Type:
        pandas.DataFrame: A DataFrame object created from the input data.
    """
    return pan.DataFrame(data)

def create_bar_chart(df):
    """
    Description:
        Creates and displays a bar chart from a Pandas DataFrame. Each column in the DataFrame is represented by a separate bar set.
        The bars are positioned side by side for each index value.

    Parameters:
        df (pandas.DataFrame): The DataFrame containing data to be plotted. Each column in the DataFrame is plotted as a separate set of bars.

    Return Type:
        None
    """
    header = list(df.columns)
    positions = np.arange(1, len(df) + 1)  
    width = 0.15  

    for i, head in enumerate(header):
        mp.bar(positions + i * width, df[head], width=width, label=head)

    mp.xlabel("Index")
    mp.ylabel("Values")
    mp.title("Bar Plot from DataFrame")
    
    mp.xticks(positions + width, df.index)
    mp.legend()

    mp.show()

def main():
    data = {
        'a': [2, 8, 5, 7, 6],
        'b': [4, 3, 4, 2, 6],
        'c': [6, 7, 4, 7, 8],
        'd': [8, 6, 4, 8, 6],
        'e': [10, 4, 3, 3, 2]
    }
    
    df = create_dataframe(data)
    create_bar_chart(df)

if __name__ == "__main__":
    main()
```

12. Write a Python program to create bar plots with error bars on the same figure.

```python

def create_bar_chart(mean_velocity, standard_velocity):
    
    """
    Description:
        Creates and displays a bar chart with error bars representing the mean velocity and its standard deviation.

    Parameters:
        mean_velocity (list of float): A list of mean velocities for different categories.
        standard_velocity (list of float): A list of standard deviations for the mean velocities, used as error bars.

    Return Type:
        None
    """
    
    positions = np.arange(len(mean_velocity))  
    width = 0.4  

    mp.bar(positions, mean_velocity, yerr=standard_velocity, capsize=5, color='skyblue', ecolor='red', edgecolor='black', width=width)

    mp.xlabel("Category")
    mp.ylabel("Velocity")
    mp.title("Bar Plot with Error Bars")

    mp.xticks(positions, ['Category 1', 'Category 2', 'Category 3', 'Category 4'])  

    mp.show()

def main():
    mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
    standard_velocity = [0.3314, 0.2278, 0.2836, 0.2645]

    create_bar_chart(mean_velocity, standard_velocity)

if __name__ == "__main__":
    main()
```

13. Write a Python program to create bar plots with errorbars on the same figure. Attach a text
label above each bar displaying men means (integer value).

```python

def create_bar_chart(mean_velocity, standard_velocity):
    
    """
    Description:
        Creates and displays a bar chart with error bars representing the mean velocity and its standard deviation. 
        Adds numerical labels above each bar indicating the mean velocity value.

    Parameters:
        mean_velocity (list of float): A list of mean velocities for different categories.
        standard_velocity (list of float): A list of standard deviations for the mean velocities, used as error bars.

    Return Type:
        None
    """
    
    positions = np.arange(len(mean_velocity))
    width = 0.4 

    bars = mp.bar(positions, mean_velocity, yerr=standard_velocity, capsize=5, color='skyblue', ecolor='red', edgecolor='black', width=width)

    mp.xlabel("Category")
    mp.ylabel("Velocity")
    mp.title("Bar Plot with Error Bars")

    mp.xticks(positions, ['Category 1', 'Category 2', 'Category 3', 'Category 4'])

    for i, bar in enumerate(bars):
        height = bar.get_height()  
        mp.text(bar.get_x() + bar.get_width() / 2, height, f'{int(mean_velocity[i])}', ha='center', va='bottom')

    mp.show()

def main():
    mean_velocity = [0.2474, 0.1235, 0.1737, 0.1824]
    standard_velocity = [0.3314, 0.2278, 0.2836, 0.2645]

    create_bar_chart(mean_velocity, standard_velocity)

if __name__ == "__main__":
    main()
```
