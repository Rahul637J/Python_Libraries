```python


def read_text_file(file_name):
    """
    Description:
        Reads data from a text file and extracts two columns of data, assuming space-separated values.
    Parameters:
        file_name (str): The name of the text file to be read.
    Returns:
        tuple: Two lists, one for the 'x' values and one for the 'y' values.
    """
    datas = pan.read_csv(file_name, delimiter=' ', header=None, names=['x', 'y'])
    return datas['x'], datas['y']

def create_plot(rows, columns):
    """
    Description:
        Creates and displays a line plot based on the given x (rows) and y (columns) values.
        The plot includes markers on data points, a title, axis labels, and specific x-ticks.
    Parameters:
        rows (list): A list of x-axis values.
        columns (list): A list of y-axis values.
    Return Type:
        None
    """
    mp.plot(rows, columns)
    mp.title("Matplot Title")
    mp.xlabel("X-axis")
    mp.ylabel("Y-axis")
    mp.xticks([1, 2, 3, 4])
    mp.show()

def main():

    file_name = 'test.txt'
    row, column = read_text_file(file_name)

    print(row)

    create_plot(row, column)

if __name__ == "__main__":
    main()

```

4. Write a Python program to draw line charts of the financial data of Alphabet Inc. between
October 3, 2016 to October 7, 2016.

```python

def read_text_file(file_name):
    """
    Description:
        Reads data from a CSV file into a Pandas DataFrame.
    Parameters:
        file_name (str): The name of the CSV file to be read.
    Returns:
        pd.DataFrame: A DataFrame containing the data from the CSV file.
    """
    datas = pan.read_csv(file_name)
    return datas

def create_plot(datas):
    """
    Description:
        Creates and displays a line plot of financial data for Alphabet Inc. between October 3, 2016,
        and October 7, 2016. The plot includes 'Open', 'High', 'Low', and 'Close' prices.
    Parameters:
        datas (pd.DataFrame): A DataFrame containing the financial data with columns 'Date', 'Open', 'High', 'Low', and 'Close'.
    Return Type:
        None
    """
    datas['Date'] = pan.to_datetime(datas['Date'], format='%d-%m-%y')

    mp.plot(datas['Date'], datas['Open'], marker='o', linestyle='-', label='Open')
    mp.plot(datas['Date'], datas['High'], marker='o', linestyle='-', label='High')
    mp.plot(datas['Date'], datas['Low'], marker='o', linestyle='-', label='Low')
    mp.plot(datas['Date'], datas['Close'], marker='o', linestyle='-', label='Close')
    
    mp.title("Financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.")
    mp.xlabel("Date")
    mp.ylabel("Price")
    mp.legend()
    mp.tight_layout()
    mp.show()

def main():
    
    file_name = 'fdata.csv'
    datas = read_text_file(file_name)

    create_plot(datas)

if __name__ == "__main__":
    main()
```

9.Write a Python program to display the current axis limits values and set new axis values.

```python

def create_plot():
    """
    Description:
        Plots a line chart with two sets of axis limits. The first plot uses the existing axis limits based on the provided data,
        while the second plot demonstrates how to change the axis limits manually. The function also prints the current axis limits
        before displaying the plots.
    Parameters:
        None
    Return Type:
        None
    """
    x = [1, 2, 3, 4, 5]
    y = [1, 4, 9, 16, 25] 
    
    # Plot with existing axis limits
    mp.plot(x, y, marker='o', linestyle='-', c='blue')

    mp.xlabel('X-axis')
    mp.ylabel('Y-axis')

    mp.xlim(min(x), max(x))
    mp.ylim(min(y), max(y))

    mp.xticks(np.arange(min(x), max(x) + 1, 1))
    
    mp.title('Line Plot with Existing Axis Limits')
    print("Existing range:", mp.xlim(), mp.ylim())
    mp.show()

    # Plot with new axis limits
    mp.plot(x, y, marker='o', linestyle='-', color='blue')
    mp.xlim([1, 6])  
    mp.ylim([1, 30]) 
    mp.xlabel('X-axis')
    mp.ylabel('Y-axis')
    mp.title('Line Plot with New Axis Limits')
    print("New range:", mp.xlim(), mp.ylim())
    mp.show()

def main():
    create_plot()

if __name__ == "__main__":
    main()
```

12. Write a Python program to create multiple types of charts

```python

def create_plot():
    
    """

    Description:
        Creates and displays a figure with three subplots arranged in a single row.
        - The first subplot is a line chart with red dashed lines.
        - The second subplot is a bar chart with blue bars.
        - The third subplot is a scatter plot with green points.
        Each subplot has its own title, x-axis label, and y-axis label.
    Parameters:
        None
    Return Type:
        None

    """
    
    # Sample data
    x = np.array([0, 1, 2, 3, 4, 5])
    y1 = np.array([0, 1, 4, 9, 16, 25])
    y2 = np.array([25, 20, 15, 10, 5, 0])
    y3 = np.random.rand(6)
    
    # Create a figure with 3 subplots (1 row, 3 columns)
    fig, (ax1, ax2, ax3) = mp.subplots(1, 3, figsize=(15, 5))
    
    # Line chart on the first subplot
    ax1.plot(x, y1, 'r--')  # Red dashed line
    ax1.set_title('Line Chart')
    ax1.set_xlabel('X-axis')
    ax1.set_ylabel('Y-axis')
    
    # Bar chart on the second subplot
    ax2.bar(x, y2, color='blue')
    ax2.set_title('Bar Chart')
    ax2.set_xlabel('X-axis')
    ax2.set_ylabel('Y-axis')
    
    # Scatter plot on the third subplot
    ax3.scatter(x, y3, color='green')
    ax3.set_title('Scatter Plot')
    ax3.set_xlabel('X-axis')
    ax3.set_ylabel('Y-axis')
    
    # Adjust layout to prevent overlap
    mp.tight_layout()
    
    # Display the plots
    mp.show()

def main():
    create_plot()

if __name__ == "__main__":
    main()
```

14. Write a Python program to display the grid and draw line charts of the closing value of
Alphabet Inc. between October 3, 2016 to October 7, 2016. Customized the grid lines with
rendering with a larger grid (major grid) and a smaller grid (minor grid).Turn on the grid but turn
off ticks.

```python

def create_plot():
    
    """

    Description:
        Creates and displays a line plot of the closing values for Alphabet Inc. 
        between October 3, 2016, and October 7, 2016. The function simulates reading
        financial data from a dictionary and plots it with customized grid lines and ticks.
        The plot includes major and minor grid lines with different styles and colors, and
        disables the tick marks on both axes.

    Parameters:
        None

    Return Type:
        None

    """
    
    # Sample data as a dictionary
    data = {
        'Date': ['03-10-16', '04-10-16', '05-10-16', '06-10-16', '07-10-16'],
        'Close': [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]
    }
    
    # Convert the dictionary to a pandas DataFrame
    df = pan.DataFrame(data)
    
    # Convert 'Date' to datetime for better plotting
    df['Date'] = pan.to_datetime(df['Date'], format='%d-%m-%y')
    
    # Plotting the Close values
    mp.plot(df['Date'], df['Close'], marker='o', linestyle='-', color='red', label='Close Value')
    
    # Customize the grid
    mp.grid(True, which='major', linestyle='-', linewidth=0.75, color='blue')
    
    # Turn on minor ticks and customize the minor grid
    mp.minorticks_on()  
    mp.grid(True, which='minor', linestyle=':', linewidth=0.5, color='gray')
    
    # Turn off ticks
    mp.tick_params(axis='both', which='both', length=0)
    
    # Add labels, title, and legend
    mp.xlabel('Date')
    mp.ylabel('Close Value')
    mp.title('Alphabet Inc. Closing Value (Oct 3, 2016 - Oct 7, 2016)')
    mp.legend()
    
    # Rotate date labels for better readability
    mp.xticks(rotation=45)
   
    # Display the plot
    mp.show()

def main():
    create_plot()

if __name__ == "__main__":
    main()
```
