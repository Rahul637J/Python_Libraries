```python
import seaborn as sns
import pandas as pan
import matplotlib.pyplot as mp
```

1. Write a program to draw a scatter plot of “day” against “total bill” for a dataset given in a
url

```python

def read_data():
    """
    Description:
        Reads the tips dataset from the given URL in CSV format and returns it as a Pandas DataFrame.
        
    Parameters:
        None
        
    Return Type:
        pd.DataFrame: A DataFrame containing the tips dataset.
    """
    data = pan.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
    return data

def create_scatter_plot(data):
    """
    Description:
        Creates and displays a scatter plot of total bill against day of the week.
        
    Parameters:
        data (pd.DataFrame): A DataFrame containing the tips dataset with 'day' and 'total_bill' columns.
        
    Return Type:
        None
    """
    
    sns.scatterplot(x='day', y='total_bill', data=data, hue='day')
    
    mp.title("Scatter Plot of Total Bill by Day")
    
    mp.show()

def main():
    data = read_data()
    
    create_scatter_plot(data)

if __name__ == "__main__":
    main()
```
