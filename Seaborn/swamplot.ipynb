```python
import seaborn as sns
import pandas as pan
import matplotlib.pyplot as mp
```

1. Write a program to draw a swarm plot of total bill against size for a given dataset

```python

def read_data():
    """
    Description:
        Reads the tips dataset from the given URL in CSV format and returns it as a Pandas DataFrame.
        
    Parameters:
        None
        
    Return Type:
        pd.DataFrame: A DataFrame containing the tips dataset.
    """
    
    data = pan.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
    return data

def create_strip_plot(data):
    """
    Description:
        Creates and displays a strip plot of total bill against size.
        
    Parameters:
        data (pd.DataFrame): A DataFrame containing the tips dataset with 'size' and 'total_bill' columns.
        
    Return Type:
        None
    """
    
    sns.stripplot(x='size', y='total_bill', data=data, hue='size')
    
    mp.title("Strip Plot of Total Bill Against Size")
    
    mp.show()

def main():
    data = read_data()
    
    create_strip_plot(data)

if __name__ == "__main__":
    main()
```
