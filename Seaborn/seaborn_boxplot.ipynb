```python
import seaborn as sns
import pandas as pan
import matplotlib.pyplot as mp
```

1. Write a program to draw box plot of life expectancy by continent for a data set given in a
url

```python

def read_data():
    """
    Description:
        Reads the Gapminder dataset from the given URL in CSV format and returns it as a Pandas DataFrame.
        
    Parameters:
        None
        
    Return Type:
        pd.DataFrame: A DataFrame containing the Gapminder dataset.
    """
    
    data = pan.read_csv("https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv")
    return data

def create_box_plot(data):
    """
    Description:
        Creates and displays a box plot of life expectancy by continent.
        
    Parameters:
        data (pd.DataFrame): A DataFrame containing the Gapminder dataset with 'continent' and 'lifeExp' columns.
        
    Return Type:
        None
    """
    
    sns.boxplot(x='continent', y='lifeExp', data=data, hue='continent')
    
    mp.title("Life Expectancy by Continent")
    
    mp.show()

def main():
    data = read_data()
    
    create_box_plot(data)

if __name__ == "__main__":
    main()
```

2. Write a program to draw a box plot of day by tips for a dataset given in a url

```python

def read_data():
    """
    Description:
        Reads the tips dataset from the given URL in CSV format and returns it as a Pandas DataFrame.
        
    Parameters:
        None
        
    Return Type:
        pd.DataFrame: A DataFrame containing the tips dataset.
    """
    
    data = pan.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
    return data

def create_box_plot(data):
    """
    Description:
        Creates and displays a box plot of tips based on the day of the week.
        
    Parameters:
        data (pd.DataFrame): A DataFrame containing the tips dataset with 'day' and 'tip' columns.
        
    Return Type:
        None
    """
    
    sns.boxplot(x='day', y='tip', data=data, hue='day')
    
    mp.title("Box Plot of Tips by Day")
    
    mp.show()

def main():
    data = read_data()
    
    create_box_plot(data)

if __name__ == "__main__":
    main()
```
