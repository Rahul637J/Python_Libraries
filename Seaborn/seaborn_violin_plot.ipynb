```python
import seaborn as sns
import pandas as pan
import matplotlib.pyplot as mp
```

1. Write a program to draw a violin plot of sex against total_bill for a given dataset

```python

def read_data():
    """
    Description:
        Reads the tips dataset from the given URL in CSV format and returns it as a Pandas DataFrame.
        
    Parameters:
        None
        
    Return Type:
        pd.DataFrame: A DataFrame containing the tips dataset.
    """
    data = pan.read_csv("https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv")
    return data

def create_violin_plot(data):
    """
    Description:
        Creates and displays a violin plot comparing the distribution of total bill amounts based on sex.
        
    Parameters:
        data (pd.DataFrame): A DataFrame containing the tips dataset with 'sex' and 'total_bill' columns.
        
    Return Type:
        None
    """

    sns.violinplot(x='sex', y='total_bill', data=data, hue='sex')
    
    mp.title("Violin Plot of Total Bill by Sex")
    
    mp.show()

def main():
    data = read_data()
    
    create_violin_plot(data)

if __name__ == "__main__":
    main()
```

2. Write a program to draw a violin plot of “species” against “sepal length” for a dataset
given in a url

```python

def read_data():
    """
    Description:
        Reads the Iris dataset from the given URL in CSV format and returns it as a Pandas DataFrame.
        
    Parameters:
        None
        
    Return Type:
        pd.DataFrame: A DataFrame containing the Iris dataset.
    """
    data = pan.read_csv("https://github.com/mwaskom/seaborn-data/raw/master/iris.csv")
    return data

def create_violin_plot(data):
    """
    Description:
        Creates and displays a violin plot comparing sepal length based on species.
        
    Parameters:
        data (pd.DataFrame): A DataFrame containing the Iris dataset with 'species' and 'sepal_length' columns.
        
    Return Type:
        None
    """
    sns.violinplot(x='species', y='sepal_length', data=data, hue='species')
    
    mp.title("Violin Plot of Sepal Length by Species")
    
    mp.show()

def main():
    data = read_data()
    
    create_violin_plot(data)

if __name__ == "__main__":
    main()
```
