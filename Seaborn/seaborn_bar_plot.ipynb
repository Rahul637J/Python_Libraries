```python
import seaborn as sns
import pandas as pan
import matplotlib.pyplot as mp
```

1. Write a program to draw bar plot of sex against survived for a dataset given in the url

```python

def read_data():
    """
    Description:
        Reads Titanic dataset from the given URL in CSV format and returns it as a Pandas DataFrame.
        
    Parameters:
        None
        
    Return Type:
        pd.DataFrame: A DataFrame containing the Titanic dataset.
    """
    data = pan.read_csv("https://github.com/mwaskom/seaborn-data/raw/master/titanic.csv")
    return data

def create_bar_plot(data):
    """
    Description:
        Creates and displays a bar plot comparing the number of survivors based on sex.
        
    Parameters:
        data (pd.DataFrame): A DataFrame containing the Titanic dataset with 'sex' and 'survived' columns.
        
    Return Type:
        None
    """
    sns.barplot(x=data['sex'], y=data['survived'], data=data, hue='sex', palette={'male':'r', 'female':'g'}, dodge=False)
    
    mp.title("The data represents sex against survival status")
    
    # Remove legend box
    mp.legend([], [], frameon=False)
    
    mp.show()

def main():
    data = read_data()
    create_bar_plot(data)

if __name__ == "__main__":
    main()
```
