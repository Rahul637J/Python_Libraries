```python
import pandas as pan
import plotly.io as pio
import plotly.graph_objs as go
import plotly.express as px
import numpy as np
```

1. Write a program to draw a scatter plot for random 1000 x and y coordinates

```python

pio.renderers.default = 'notebook_connected'

def create_scatter_plot(x, y):
    fig = go.Figure(data=go.Scatter(x=x, y=y, mode='markers', marker=dict(size=5, opacity=0.5)))
    
    fig.update_layout(
        title="Scatter Plot of Random Coordinates",
        xaxis_title="X Coordinate",
        yaxis_title="Y Coordinate",
        template="plotly_white"
    )
    
    fig.show()

def main():
    num_points = 1000
    x = np.random.randint(low=1,high=1000,size=num_points)
    y = np.random.randint(low=1,high=1000,size=num_points)
    
    create_scatter_plot(x, y)

if __name__ == "__main__":
    main()
```

2. Write a program to draw line and scatter plots for random 100 x and y coordinates

```python

def create_line_and_scatter_plots(x, y):
    """
    Creates and displays both a line plot and a scatter plot of x vs y coordinates using Plotly.

    Parameters:
        x (np.ndarray): Array of x coordinates.
        y (np.ndarray): Array of y coordinates.

    Return Type:
        None
    """

    fig = go.Figure()

    fig.add_trace(go.Scatter(x=x, y=y, mode='markers', name='Scatter Plot', marker=dict(size=8, color='rgba(219, 64, 82, 0.8)')))

    fig.add_trace(go.Scatter(x=x, y=y, mode='lines', name='Line Plot', line=dict(color='royalblue', width=2)))

    fig.update_layout(
        title="Line and Scatter Plots of Random Coordinates",
        xaxis_title="X Coordinate",
        yaxis_title="Y Coordinate",
        template="plotly_white"
    )

    fig.show()

def main():
    num_points = 100
    x = np.random.randint(low=1,high=100,size=num_points)
    y = np.random.randint(low=1,high=100,size=num_points)
    print(x)
    
    create_line_and_scatter_plots(x, y)

if __name__ == "__main__":
    main()
```

3. Write a program to draw a scatter plot for random 500 x and y coordinates and style it

```python

def create_styled_scatter_plot(x, y):
    
    """
    Creates and displays a styled scatter plot of x vs y coordinates using Plotly.

    Parameters:
        x (np.ndarray): Array of x coordinates.
        y (np.ndarray): Array of y coordinates.

    Return Type:
        None
    """
    fig = go.Figure()

    # Add scatter plot with custom styling
    fig.add_trace(go.Scatter(
        x=x,
        y=y,
        mode='markers',
        marker=dict(
            size=12,  # Size of markers
            color=np.random.rand(len(x)),  # Marker color based on random values
            colorscale='Viridis',  # Color scale for markers
            showscale=True,  # Show color scale
            line=dict(
                width=2,  # Border width around markers
                color='rgba(255, 0, 0, 0.5)'  # Border color around markers
            )
        ),
        text=['Point {}'.format(i) for i in range(len(x))],  # Hover text
        hoverinfo='text'
    ))

    # Update layout with custom styling
    fig.update_layout(
        title="Styled Scatter Plot of Random Coordinates",
        xaxis_title="X Coordinate",
        yaxis_title="Y Coordinate",
        template="plotly_dark",  # Dark theme for the plot
        title_font=dict(size=24, family='Arial', color='white'),  # Title font styling
        xaxis_title_font=dict(size=18, family='Arial', color='white'),  # X-axis title font styling
        yaxis_title_font=dict(size=18, family='Arial', color='white'),  # Y-axis title font styling
        paper_bgcolor='black',  # Background color of the paper
        plot_bgcolor='black'  # Background color of the plot area
    )

    fig.show()

def main():
    num_points = 500
    x = np.random.rand(num_points)
    y = np.random.rand(num_points)
    
    create_styled_scatter_plot(x, y)

if __name__ == "__main__":
    main()
```

4. Write a program to draw a scatter plot for a given dataset and show datalabels on hover

```python

def read_data():
    """
    Reads the dataset from the given URL and returns it as a Pandas DataFrame.

    Parameters:
        None

    Return Type:
        pd.DataFrame: A DataFrame containing the dataset with columns like 'State', 'Population', and 'Area'.
    """
    url = "https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv"
    data = pan.read_csv(url)
    return data

def create_scatter_plot(data):
    """
    Creates and displays a scatter plot of states' population with data labels on hover.

    Parameters:
        data (pd.DataFrame): A DataFrame containing columns 'State', 'Population'.

    Return Type:
        None
    """
    fig = px.scatter(
        data_frame=data,
        x='Rank', 
        y='Population',
        text='State',  
        title="Scatter Plot of USA States by Population and Rank",
        labels={'Rank': 'Rank', 'Population': 'Population'},
        template='plotly_white'  
    )

    fig.update_traces(textposition='top center', marker=dict(size=10, color='rgba(255, 0, 0, 0.8)', line=dict(width=2, color='rgba(255, 0, 0, 0.8)')))
    fig.update_layout(
        xaxis_title="Rank",
        yaxis_title="Population",
        title_font=dict(size=24, family='Arial', color='black'),
        xaxis_title_font=dict(size=18, family='Arial', color='black'),
        yaxis_title_font=dict(size=18, family='Arial', color='black'),
        paper_bgcolor='white',
        plot_bgcolor='white'
    )

    fig.show()

def main():
    data = read_data()
    create_scatter_plot(data)

if __name__ == "__main__":
    main()
```
